## Lemmatizer

```python
wordnet = WordNetLemmatizer()
```

```python
corp=[ads.loc[x,'ad_copy'] for x in range(len(ads))]
```

```python
ads[~ads['ad_copy'].isna()]
```

```python
lemmed_corp=[]
for word in range(len(corp)):
    if corp[word]!= None:
        lemmed_corp.append(wordnet.lemmatize(corp[word]))
```

## TFIDF

```python
tf = TfidfVectorizer(stop_words='english')
```

```python
corp_vect = tf.fit_transform(lemmed_corp)
```

## Count words/ Vocab list

```python
count = CountVectorizer(stop_words='english')
counted_corp = count.fit_transform(lemmed_corp)
```

```python
count.
```

```python
sum_words = counted_corp.sum(axis=0)
```

```python
words_freq = [(word, sum_words[0,idx]) for word, idx in count.vocabulary_.items()]
```

```python
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
```

## Dummies for Interests

```python
interest_set = set()
ads['interests_categories'].values
```

```python
for lst in ads['interests_categories']:
    for cat in lst:
        interest_set.add(cat)
```

```python
dums = ads.copy(deep=True)
```

```python
for cat in interest_set:
    for row in range(dums.shape[0]):
        if cat in dums.loc[row,'interests_categories']:
            dums.loc[row,cat]=int(1)
        else:
            dums.loc[row,cat]=int(0)
```

```python
for cat in interest_set:
    dums[cat] = dums[cat].astype(int)
```

```python
target_interest = dums[list(interest_set)]
```

### At this point, it is important to work only with ads with ad_copy text present, so X, y shapes match up

```python
target_interest = dums[~dums['ad_copy'].isna()][list(interest_set)]
```
